```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Select the specified columns
selected_columns = [
    'IQ', 'education_expenditure', 'avg_income', 'avg_temp', 'region',
    'happiness_score', 'gdp_per_capita', 'social_support',
    'healthy_life_expectancy', 'freedom_to_make_life_choices',
    'generosity', 'perceptions_of_corruption'
]

# Ensure 'region' is excluded from correlation calculation as it is categorical
numeric_columns = [
    'IQ', 'education_expenditure', 'avg_income', 'avg_temp',
    'happiness_score', 'gdp_per_capita', 'social_support',
    'healthy_life_expectancy', 'freedom_to_make_life_choices',
    'generosity', 'perceptions_of_corruption'
]

# Select only the numeric columns for correlation calculation
numeric_df = iq_happy_data[numeric_columns]

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Display the correlation matrix
print(correlation_matrix)

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## Do countries that spend more on education have happier citizens, or just students buried under more homework? (Investigate how education expenditure correlates with happiness)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with 'education_expenditure' and 'happiness_score' columns

# Select relevant columns and drop missing values
data = iq_happy_data[['education_expenditure', 'happiness_score']].dropna()

# Calculate the correlation coefficient
correlation = data['education_expenditure'].corr(data['happiness_score'])
print(f'Correlation coefficient between education expenditure and happiness score: {correlation:.2f}')

# Create a scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='education_expenditure', y='happiness_score', data=data)
plt.title('Education Expenditure vs. Happiness Score')
plt.xlabel('Education Expenditure (% of GDP)')
plt.ylabel('Happiness Score')
plt.show()
```

## Are people in colder climates less happy, or do they just have a cooler attitude? (Analyze the impact of average temperature on happiness) 

```python
import plotly.express as px

# Assuming iq_happy_data is your DataFrame

fig = px.scatter(
    iq_happy_data,
    x='avg_temp',
    y='happiness_score',
    color='region',
    size='gdp_per_capita',  # Size of markers based on GDP per capita
    hover_name='country',
    title='Average Temperature vs. Happiness Score by Region',
    labels={'avg_temp': 'Average Temperature (°C)', 'happiness_score': 'Happiness Score'}
)
fig.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming iq_happy_data is your DataFrame with 'avg_temp' and 'happiness_score' columns

# Select relevant columns and drop missing values
data = iq_happy_data[['avg_temp', 'happiness_score']].dropna()

# Calculate the correlation coefficient
correlation = data['avg_temp'].corr(data['happiness_score'])
print(f'Correlation coefficient between average temperature and happiness score: {correlation:.2f}')

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='avg_temp', y='happiness_score', data=data, scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
plt.title('Average Temperature vs. Happiness Score')
plt.xlabel('Average Temperature (°C)')
plt.ylabel('Happiness Score')
plt.show()

# Heatmap of the correlation matrix
corr_matrix = data.corr()
plt.figure(figsize=(5, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

# Joint plot with regression line
sns.jointplot(x='avg_temp', y='happiness_score', data=data, kind='reg', height=8, scatter_kws={'alpha': 0.6})
plt.xlabel('Average Temperature (°C)')
plt.ylabel('Happiness Score')
plt.show()
```
